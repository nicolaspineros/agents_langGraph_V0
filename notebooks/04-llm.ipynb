```python
import os

from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt_4o = ChatOpenAI(model="gpt-4o-mini", temperature=0)
response = gpt_4o.invoke("Hello, how are you?")
response.pretty_print()
```

```python
gpt35 = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
response = gpt35.invoke("Hello, how are you?")
response.pretty_print()
```

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in spanish")
msg_2 = HumanMessage(content="Me llamo Nicolas")
msg_3 = AIMessage(content="Hola Nicolas, ¿cómo estás?")
msg_4 = HumanMessage(content="Como me llamo?")
history = [msg_1, msg_2, msg_3, msg_4]
response = gpt_4o.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(model="gpt-4o-mini", temperature=0)
response = llm.invoke(history)
response.pretty_print()
```
