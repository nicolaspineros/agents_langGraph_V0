```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(model="gpt-4o-mini", temperature=0)
```

```python
openai_vector_store_ids = ["vs_691c8eeb51988191bf2286f5cd54ffbb"]

file_search_tool = {
    "type": "file_search",
    "vector_store_ids": openai_vector_store_ids,
}

# Agrega la herramienta de búsqueda de archivos al modelo
llm_with_tools = llm.bind_tools([file_search_tool])
```

```python
response = llm_with_tools.invoke("Como mejoro el rendimiento de mi website")
response.text
```
