```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model(model="gpt-4o-mini", temperature=0)

# Damos datos falsos por seguridad
response = llm.invoke("Hola, soy Nicolas y mi telefono es 3178654321")
response.text
```

```python
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):
    name: str = Field(description="The name of the person")
    phone: str = Field(description="The phone number of the person")
    email: str = Field(description="The email of the person")
    tone: int = Field(description="The tone of the person", ge=0, le=100)
    age: int = Field(description="The age of the person")
    sentiment: str = Field(description="The sentiment of the conversation")

llm_with_structured_output = llm.with_structured_output(ContactInfo)

messages = [
    ("system", "You are a helpful assistant that extracts information from a given conversation"),
    ("user", "Hola, soy Nicolas y mi telefono es 3178654321"),
]

response = llm_with_structured_output.invoke(messages)
response
```
